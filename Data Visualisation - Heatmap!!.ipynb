```python
yse = sns.load_dataset('flights')
yse = yse.pivot('month', 'year', 'passengers')
display_yse = sns.heatmap(yse)
plt.savefig('Image01.png', dpi=300)
```

```python
display_yse = sns.heatmap(yse, annot=True, fmt='d')
plt.savefig('Image02.png', dpi=300)
```

```python
display_yse = sns.heatmap(yse, linewidth=2)
plt.savefig('Image03.png', dpi=300)
```

```python
label = ['Doc', 'Cat', 'Wolf', 'Lion']
sizes = [50,45,60,80]

#Creating the pie chart
plt.pie(sizes, labels = label,autopct='%1.1f%%', shadow=True, startangle = 90,explode = (0.1,0.1,0,0))
plt.title("Pie Chart Demo")
plt.savefig('Image08.png', dpi=300)
plt.show()
```

How to create a Doughnut Chart

```python
#define the dataset
group_names = ["GroupeA", "GroupeB", "GroupeC"]
group_size = [20,30,50]
size_centre = [5]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#creation the doughout chart
pie1 = plt.pie(group_size,labels = group_names, radius = 1.5,colors = [a(0.5),b(1.5),c(1.0)])
pie2 = plt.pie(size_centre,radius = 1,colors ='w')
plt.title("Pie Doughout Chart Demo")
plt.savefig('Image09.png', dpi=300)
plt.show()
```
